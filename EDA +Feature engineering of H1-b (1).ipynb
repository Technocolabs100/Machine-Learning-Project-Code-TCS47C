```python
def wage_categorization(wage):
    if wage <=50000:
        return "VERY LOW"
    elif wage >50000 and wage <= 70000:
        return "LOW"
    elif wage >70000 and wage <= 90000:
        return "MEDIUM"
    elif wage >90000 and wage<=150000:
        return "HIGH"
    elif wage >=150000:
        return "VERY HIGH"
```

```python
df1['WAGE_CATEGORY'] = df1['PREVAILING_WAGE_1'].apply(wage_categorization)
df1['WAGE_CATEGORY']
```
